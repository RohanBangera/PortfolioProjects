# Spotify EDA

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_songs = pd.read_csv('C:/Users/rohan/Documents/SpotifyTracks.csv')
df_songs.head()
```

```python
pd.isnull(df_songs).sum()
```

```python
sorted_df = df_songs.sort_values('popularity')
sorted_df.head()
```

```python
df_songs.describe().transpose()
```

```python
most_popular = df_songs.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_songs.set_index("release_date",inplace = True)
df_songs.index = pd.to_datetime(df_songs.index)
df_songs.head()
```

```python
df_songs["duration"] = df_songs["duration_ms"].apply(lambda x:round(x/1000))
df_songs.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_songs.head()
```

```python
corr_df = df_songs.drop(["key", "mode","time_signature"],axis = 1).corr(method = "pearson")
plt.figure(figsize=(15,7))
heatmap = sns.heatmap(corr_df, annot = True, fmt ='0.1g', vmin = -1, vmax = 1, center = 0, cmap = "inferno",linewidths = 1, linecolor = "black")
heatmap.set_title('Correlation heatmap between all variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_df = df_songs.sample(int(0.01*len(df_songs)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize =(10,6))
sns.regplot(data =sample_df, y = "loudness", x ="energy", color = "c" ).set(title = "correlation between loudness vs energy")
```

```python
df_songs['dates'] = df_songs.index.get_level_values('release_date')
df_songs.dates = pd.to_datetime(df_songs.dates)
years = df_songs.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect = 2, height =4, kind = 'hist').set(title= "Number of songs per year")
```

```python
total_dur = df_songs.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig =sns.barplot(x = years, y = total_dur, ax =ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```

```python
df_genre = pd.read_csv('C:/Users/rohan/Documents/SpotifyFeatures.csv')
```

```python
plt.title("Avg Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap =True)
sns.barplot(y = 'genre', x = 'duration_ms',data = df_genre)
plt.xlabel("duration in milliseconds")
plt.ylabel("Genres")
```
